# Question 4: Implementing a Multinomial Naive Bayes Classifier

We need to implement the multinomial naive bayes training algorithm

1) We need to extract the vocabulary from the traindata.txt

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer

data = pd.read_csv('traindata.txt', sep='\t', header = None)
data.head()
```

```python
trainingData = []
```

```python
for i in range(len(data)):
    test = data.iloc[i,0]
    trainingData.append(test.split())

trainingData
```

```python
data = pd.DataFrame(trainingData)
data
```

```python
labels = pd.read_csv('trainlabels.txt', sep='\t', header = None)
labels
```

```python
labels.columns = ['class']
pd.concat([data, labels], axis=1)
```

```python
text_data = np.array(['I love Brazil. Brazil!',
                      'Brazil is best',
                      'Germany beats both'])
```

```python
# Create bag of words
count = CountVectorizer()
bag_of_words = count.fit_transform(text_data)

# Create feature matrix
X = bag_of_words.toarray()
```

```python
y = np.array([0,0,1])
```

```python
# Create multinomial naive Bayes object with prior probabilities of each class
clf = MultinomialNB(class_prior=[0.25, 0.5])

# Train model
model = clf.fit(X, y)
```

```python
# Create new observation
new_observation = [[0, 0, 0, 1, 0, 1, 0]]
```

```python
# Predict new observation's class
model.predict(new_observation)
```
